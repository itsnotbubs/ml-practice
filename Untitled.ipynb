```python
import numpy as np
import pandas as pd
import scipy as sp
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
#as_frame = True is a short cut to converting this loaded csv into a dataframe object
cancer = load_breast_cancer(as_frame = True)
```

```python
cancer_df = cancer.frame
cancer_df.shape
```

```python
cancer_df.head()
```

```python
cancer_df.target.value_counts()
```

```python
cancer_df.target.value_counts(normalize=True)
```

```python
cancer_features = cancer_df.drop(columns='target')
cancer_features.columns
```

```python
# using vert=False makes the box plot horizontal
cancer_features.boxplot(vert=False)
```

```python
import matplotlib.pyplot as plt
# create a grid of plots, with 5 rows and 6 columns
# and place it on a canvas 18 (virtual) inches whide and 6 inches high
fig, axes = plt.subplots(5, 6, figsize=(18, 6))
# for each column (and with one axis after another) since we have 30 plots .ravel() flatten the 5x6 into a array
for c, ax in zip(cancer_features.columns, axes.ravel()):
    # create a box-plot grouped by the target variabel
    cancer_df[[c, 'target']].boxplot(vert=False, by='target', ax=ax)
    # remove the x-axis label
    ax.set_xlabel("")
# remove the figure title
plt.suptitle("")
# fit everything nicely on the plot
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
data_train, data_test = train_test_split(cancer_df)
```

```python
data_train.shape
```

```python
data_test.shape
```

```python
X_train = data_train.drop(columns='target')
y_train = data_train.target
X_test = data_test.drop(columns='target')
y_test = data_test.target
```

```python
# the return value is the axes that was created for the plot
ax = X_train.plot.scatter(x='mean compactness', y='worst concave points',
                     # use the target to color points
                     # use the blue-white-red colormap
                     c=y_train, colormap='bwr', s=2)
ax.set_xlabel("mean compactness")
ax.set_ylabel("worst concave points")
# We set the aspect ratio to equal to faithfully represent distances between points.
ax.set_aspect("equal")
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
# Here, we set the number of neighbors to one
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
# because fit returns self, Jupyter renders a representation of the object
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
y_pred.shape
```

```python
(y_pred == y_test).mean()
```

```python
print("accuracy: ", knn.score(X_test, y_test))
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
print("accuracy: ", knn.score(X_test, y_test))
```
